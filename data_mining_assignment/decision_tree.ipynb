```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Load dataset
file_path = "D:/dm_assignment/datasets/decision_tree_dataset.csv"
df = pd.read_csv(file_path)

# Separate features and target
X = df[['Temp', 'Taste', 'Size']]
y = df['Appeal']
```

```python
# Encode categorical variables
encoders = {}
for column in X.columns:
    enc = LabelEncoder()
    X[column] = enc.fit_transform(X[column])
    encoders[column] = enc

y_enc = LabelEncoder().fit_transform(y)

# Train Decision Tree
model = DecisionTreeClassifier(criterion="entropy", random_state=0)
model.fit(X, y_enc)
```

```python
# Plot Decision Tree
plt.figure(figsize=(12,6))
plot_tree(model, feature_names=X.columns, class_names=['No', 'Yes'], filled=True, rounded=True)
plt.title("Decision Tree for Appeal Prediction")
plt.show()
```
